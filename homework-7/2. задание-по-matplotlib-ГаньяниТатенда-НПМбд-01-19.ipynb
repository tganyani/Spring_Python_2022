```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special
```

```python
#111111111111111111111111111111111111
#задание 1.1
f_1 = lambda x: (x**2)*np.sin(1/x**2) + x
x_1 = np.linspace(-1,1,250)
```

```python
plt.plot(x_1,f_1(x_1))
plt.xlabel('$X$')
plt.ylabel('$f(X)$')
plt.legend(['$x^2sin\\frac{1}{x^2}+x$'])
plt.title('$Exercise1\ задач 1$')
```

```python
#задание 1.2
xx = np.linspace(5,25,250)
yy = abs(1 - (1 / (1 + xx**2)) / (1 / (xx**2)))
fig, ax = plt.subplots()
ax.semilogy(xx, yy, label = '$1-\\frac{\\frac{1}{1+x^2}}{\\frac{1}{x^2}}$')
ax.set_xlabel('x')
ax.set_xlabel('y')
ax.legend(loc ='upper left')
```

```python
xx = np.linspace(-10,10,100)
yy1 = special.gamma(xx)
yy2 = special.wofz(xx)
yy3 = special.lambertw(xx)
fig, ax = plt.subplots(1, 2)
ax[0].plot(xx, yy1, label = 'gamma function')
ax[0].set_xlabel('$x$')
ax[0].set_ylabel('$y$')
ax[0].legend()
ax[1].plot(xx, yy2, color = 'red', ls = '-.', label = 'Faddeeva function')
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('$y$')
ax[1].legend()
```

```python
x = np.linspace(-15, 5, 201)
ai, aip, bi, bip = special.airy(x)
```

```python
plt.plot(x, ai, 'r', label='$Ai(x)$')
plt.plot(x, bi, 'b--', label='$Bi(x)$')
plt.ylim(-0.5, 1.0)
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
#22222222222222222222222222222222222222
#задание 2
x2 = np.linspace(-6*np.pi,6*np.pi,500)
f2 = lambda x: np.sin(x)/x
x_tic = np.linspace(-6*np.pi,6*np.pi,24)
```

```python
def f_tic(x):
    y = 2*x/np.pi
    temp =[]
    for i in range(len(x)):
        if np.round(y[i]) % 2  ==0  and y[i]!=0:
            k = int(y[i]/2)
            temp.append(f'${k}'+'\pi$')
        elif y[i] ==0:
            temp.append(f'$0$')
            print(y[i])
        else:
            temp.append('$\\frac{'+f'{-2*6 + i}'+'}'+'{'+f'{2}'+'}\pi$')
    return temp
```

```python

plt.plot(x2, f2(x2),label = '$\\frac{sinx}{x}$')
plt.xlabel('$X$')
plt.ylabel('$f(X)$')
plt.xticks(x_tic,f_tic(x_tic))
plt.grid()
plt.legend(loc ='upper left')
```

```python
#3333333333333333333333333333333
#задание 3.1
f3 = lambda x,y: np.sin(x)*np.sin(y)
x_3 = y_3 = np.linspace(-3,3)
x3, y3 = np.meshgrid(x_3, y_3)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x3,y3, f3(x3,y3), 50,cmap='viridis' )
ax.set_xlabel('$X$')
ax.set_ylabel('$Y$')
ax.set_zlabel('$f(X,Y)$');
plt.title('$sin(X)sin(Y)$')
```

```python
#задание 3.2
f4 = lambda x,y:np.exp(-x**2-y**2)
x_4 = y_4 = np.linspace(-3,3)
x4, y4 = np.meshgrid(x_4, y_4)
```
